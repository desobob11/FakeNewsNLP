```python
import os
import pandas as pd
from textblob import *
import nltk
fake = pd.read_csv("ISOT/True.csv")

```

```python
    



#fake["title"] = fake.apply(lambda row: ''.join([i for i in row["title"] if ord(i) <= 127]), axis=1)
fake["title_sent"] = fake.apply(lambda row: TextBlob(row["title"]).sentiment, axis=1)
fake["title_pol"] = fake.apply(lambda x: x["title_sent"].polarity, axis=1)
fake["title_sub"] = fake.apply(lambda x: x["title_sent"].subjectivity, axis=1)
fake




```

```python
fake["body_spent"] = fake.apply(lambda row: TextBlob(row["text"]).sentiment, axis=1)
fake["body_pol"] = fake.apply(lambda x: x["body_spent"].polarity, axis=1)
fake["body_sub"] = fake.apply(lambda x: x["body_spent"].subjectivity, axis=1)
fake
```

```python
fake = fake.drop(["title_sent", "body_spent"], axis=1)
fake = fake.drop(["title", "text", "subject", "date"], axis=1)
fake["fake"] = 0
fake.to_csv("TrueProcessed.csv")
```
