```python
import os
import pandas as pd
from textblob import *
import nltk
import numpy as np
import openpyxl
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from textblob import *



data = pd.read_csv("tftest.csv")
data



```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from nltk.corpus import stopwords

stop_words = stopwords.words("english")
corpus = data["text"]
vectorizer = TfidfVectorizer(strip_accents="ascii", lowercase=True, stop_words=stop_words, max_features=None)
features = vectorizer.fit_transform(corpus).toarray()
names = vectorizer.get_feature_names_out()

feature_frame = pd.DataFrame(features, columns=names)
feature_frame


data = pd.concat([data, feature_frame], axis=1)

data


```
