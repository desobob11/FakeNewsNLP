```python
import os
import pandas as pd
from textblob import *
import nltk
import numpy as np
from sklearn import tree
import openpyxl
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
from textblob import *

df = pd.read_csv(r"data/LIAR/train.csv")



df
```

```python
df = df.drop([i for i in df.columns if i != "statement"], axis=1)

#df["tb_sent"] = df.apply(lambda row: TextBlob(row["statement"]).sentiment, axis=1)
df["tb_pol"] = df.apply(lambda x: TextBlob(x["statement"]).polarity, axis=1)
df["tb_sub"] = df.apply(lambda x: TextBlob(x["statement"]).subjectivity, axis=1)
df
```

```python
analyzer = SentimentIntensityAnalyzer()

df["vader_scores"] = df.apply(lambda x: analyzer.polarity_scores(x["statement"]), axis=1)
df["vader_neg"] = df.apply(lambda x: x["vader_scores"]["neg"], axis=1)
df["vader_neu"] = df.apply(lambda x: x["vader_scores"]["neu"], axis=1)
df["vader_pos"] = df.apply(lambda x: x["vader_scores"]["pos"], axis=1)
df = df.drop(["vader_scores"], axis=1)
```

```python
df.to_csv("textblob_vader.csv", index=False)
```
