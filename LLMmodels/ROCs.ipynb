```python
import pandas as pd
import matplotlib.pyplot as plt

# Load GPT_ISOT data
roc_data_gpt = pd.read_csv('roc_data_gpt_isot.csv')
with open('roc_auc_gpt_isot.txt', 'r') as f:
    roc_auc_gpt = float(f.read())

# Load BERT_ISOT data
roc_data_bert = pd.read_csv('roc_data_bert_isot.csv')
with open('roc_auc_bert_isot.txt', 'r') as f:
    roc_auc_bert = float(f.read())

# Load SBERT_ISOT data
roc_data_sbert = pd.read_csv('roc_data_sbert_isot.csv')
with open('roc_auc_sbert_isot.txt', 'r') as f:
    roc_auc_sbert = float(f.read())



# Plot ROC curves
plt.figure()
plt.plot(roc_data_gpt['fpr'], roc_data_gpt['tpr'], color='blue', lw=2, label=f"GPT_ISOT (area = {roc_auc_gpt:.2f})")
plt.plot(roc_data_bert['fpr'], roc_data_bert['tpr'], color='green', lw=2, label=f"BERT_ISOT (area = {roc_auc_bert:.2f})")
plt.plot(roc_data_sbert['fpr'], roc_data_sbert['tpr'], color='orange', lw=2, label=f"SBERT_ISOT (area = {roc_auc_sbert:.2f})")
plt.plot([0, 1], [0, 1], color='red', lw=2, linestyle='--')  # Diagonal line

# Customize plot
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Combined Receiver Operating Characteristic (ROC) Curves')
plt.legend(loc="lower right")
plt.show()
```
